## Probability Distribution

```python
import seaborn as sns
from scipy.stats import norm, uniform, expon

normal_data = np.random.normal(0, 1, 1000)
uniform_data = np.random.uniform(-2, 2, 1000)
exponential_data = np.random.exponential(1, 1000)

sns.kdeplot(normal_data, label="Normal", color="blue")
sns.kdeplot(uniform_data, label="Uniform", color="green")
sns.kdeplot(exponential_data, label="Exponential", color="red")

plt.title("Comparison of Different Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

## Hypothesis Testing

## Chi-Square Test

```python
from scipy.stats import chi2_contingency

data = [[50, 30, 20], # High School
 [40, 50, 10], # Bachelor's Degree
 [10, 20, 30]] # Master's Degree

chi, p,dof,expected = chi2_contingency(data)

print(chi)
print(p)
print(dof)
print(expected)

if p < 0.5:
    print("Reject the null hypo")
else:
    print("accept")
```

```python
from scipy.stats import f_oneway


group1 = [85, 87, 90, 85, 88]
group2 = [78, 75, 80, 82, 79]
group3 = [91, 93, 95, 92, 94]

f_1 , p = f_oneway(group1,group2,group3)

print(f_1,p)

if p < 0.5:
    print("Reject the null hypo")
else:
    print("accept")
```
