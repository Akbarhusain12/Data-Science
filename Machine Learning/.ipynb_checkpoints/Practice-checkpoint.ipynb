```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Files/house_prices.csv")
```

```python
df
```

```python
plt.scatter(df['House Size'],df['House Price'], marker='*', color='Red')
plt.show()
```

```python
lnr = linear_model.LinearRegression()
```

```python
lnr.fit(df[['House Size']],df[['House Price']])
```

```python
lnr.coef_
```

```python
lnr.intercept_
```

```python
lnr.predict([[3300]])
```

```python
164.24802111 * 3300 + 112242.74406332
```

```python
plt.scatter(df['House Size'],df['House Price'], marker='*', color='Red')

plt.plot(df['House Size'], lnr.predict(df[['House Size']]), color="Blue")

plt.show()
```
