## KNN - Algorithm

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix 
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Files/iris_data.csv')
```

```python
df
```

```python
c = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
d = df["species"]
```

```python
X_train2, X_test2, Y_train2, Y_test2 =  train_test_split(c,d, test_size=0.3, random_state=43)
```

```python
Scaler = StandardScaler()
```

```python
X_train2 = Scaler.fit_transform(X_train2)
```

```python
X_test2 = Scaler.transform(X_test2)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train2,Y_train2)
```

```python
y_pred2 = knn.predict(X_test2)
```

```python
print(confusion_matrix(Y_test2,y_pred2))
```

## Decision Tree Algorithm

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("Files/iris_data.csv")
df
```

```python
E = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
F = df["species"]
```

```python
X_train3, X_test3, Y_train3, Y_test3 =  train_test_split(E,F, test_size=0.3, random_state=42)
```

```python
dnn = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
```

```python
dnn.fit(X_train3,Y_train3)
```

```python
y_pred3 = dnn.predict(X_test3)
```

```python
y_pred3
```

```python
print(confusion_matrix(Y_test3,y_pred3))
```

## K-Means Algorithm

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
S = np.random.rand(100, 2)
```

```python
Kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
Kmeans.fit(S)
```

```python
centroids = Kmeans.cluster_centers_
```

```python
labels = Kmeans.labels_
```

```python
plt.scatter(S[:, 0], S[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red')
plt.title('K-Means Clustering')
plt.show()
```

## Chi2_testing

```python
from scipy.stats import chi2_contingency
```

```python
data = [[50, 30, 20], # High School
 [40, 50, 10], # Bachelor's Degree
 [10, 20, 30]]
```

```python
chi_2, p, dof, expected = chi2_contingency(data)
print(chi_2)
print(p)
print(dof)
print(expected)

if p < 0.05:
    print("reject the null hypothesis")
else:
    print("Accepet the null hypothesis")
```

## Anova Testing

```python
import numpy as np
from scipy.stats import f_oneway
```

```python
group1 = [85, 87, 90, 85, 88]
group2 = [78, 75, 80, 82, 79]
group3 = [91, 93, 95, 92, 94]
```

```python
f_stats, p = f_oneway(group1,group2,group3)

print(f_stats)
print(p)

if p < 0.05:
    print("reject the null hypothesis")
else:
    print("Accepet the null hypothesis")
```
