```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, uniform, expon, f_oneway
```

```python
normal_data = np.random.normal(0,1,1000)
uniform_data = np.random.uniform(-2,2,1000)
exponential_data = np.random.exponential(1,1000)

sns.kdeplot(normal_data, label="Normal", color="blue")
sns.kdeplot(uniform_data, label="Uniform", color="green")
sns.kdeplot(exponential_data, label="Exponential", color="red")

plt.title("Comparison of Different distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
from scipy.stats import chi2_contingency
```

```python
data = [[50,30,20],
       [40,50,10],
       [10,20,30]]

chi2, p, dof, expected = chi2_contingency(data)

print("Chi-Square Statistics : ", chi2)
print("P-value : ", p)
print("Degree of freedon : ", dof)

if p < 0.05:
    print("Reject the null hypothesis : Education level and voting preference are dependent")
else:
    print("Fail to reject the null hypothesis : Education level and voting preference are independent")
```

```python
group1 = [85,87,90,85,88]
group2 = [78,75,80,82,79]
group3 = [91,93,95,92,94]

f_stats, p_value = f_oneway(group1,group2,group3)

print("F-Statistics : ", f_stats)
print("Probability value : ", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis : Education level and voting preference are dependent")
else:
    print("Fail to reject the null hypothesis : Education level and voting preference are independent")
```
